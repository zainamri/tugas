```python
feature = 'OWN_CAR_AGE'
binning=[-float("inf"), 5, 15, 50, float("inf")]
quantiles = [0, 0.25, 0.5, 0.75, 1]

bin_feature = pd.cut(data[train_mask][feature], bins=5).values.add_categories('Nan').fillna('Nan')
data_woe_iv = (pd.crosstab(bin_feature,data[train_mask][target],normalize='columns')
             .assign(woe=lambda datax: np.log(datax[1] / datax[0]))
             .assign(iv=lambda datax: np.sum(datax['woe']*
                                           (datax[1]-datax[0]))))

data_woe_iv[['woe','iv']]
```

```python
null_num_feature = data[list_column_nan].select_dtypes(include='number').columns
```

```python
quantiles = 10
for i in null_num_feature:
#     print(i)
    try:
        bin_feature = pd.qcut(data[train_mask][i], quantiles, duplicates='drop').values.add_categories('Nan').fillna('Nan')
        data_woe_iv = (pd.crosstab(bin_feature,data[train_mask][target],normalize='columns')
                 .assign(woe=lambda datax: np.log(datax[1] / datax[0]))
                 .assign(iv=lambda datax: np.sum(datax['woe']*
                                               (datax[1]-datax[0]))))
        data[f'{i}_WOE'] = pd.qcut(data[i], quantiles, duplicates='drop', labels=list(data_woe_iv['woe'])[:-1])
        data[f'{i}_WOE'] = data[f'{i}_WOE'].values.add_categories('Nan').fillna('Nan')
        data[f'{i}_WOE'] = data[f'{i}_WOE'].replace('Nan',list(data_woe_iv['woe'])[-1])
        data[f'{i}_WOE'] = data[f'{i}_WOE'].astype(float)
    except Exception as e:
        continue
```

## Logistic Regression

```python
from sklearn import metrics
from sklearn.model_selection import cross_val_score
from sklearn.metrics import roc_auc_score

def stepwise_logistic_regression(data, target, significance_level=0.05):
    """Perform stepwise logistic regression."""
    initial_features = final_features
    included = list(initial_features)
    while True:
        changed = False
        # Fit the model with current features
        formula = f"{target} ~ " + " + ".join(included)
        model = logit(formula, data=data).fit(disp=0)

        # Get p-values
        p_values = model.pvalues[1:]  # Exclude the intercept
        # Find the feature with the highest p-value
        max_p_value = p_values.max()
        if max_p_value > significance_level:
            # Remove the feature with the highest p-value
            excluded_feature = p_values.idxmax()
            included.remove(excluded_feature)
            changed = True
        else:
            break  # Exit if no features can be removed

    return model, included
```

```python
# Perform stepwise logistic regression
LogReg, selected_features = stepwise_logistic_regression(data[train_mask][final_features + [target]], target)
```

```python
# Print summary of the final model
train_pred_lr = LogReg.predict(data[train_mask][final_features])
test_pred_lr = LogReg.predict(data[test_mask][final_features])
print("\nROC AUC Train:",roc_auc_score(data[train_mask][target],train_pred_lr))
print("ROC AUC Test:",roc_auc_score(data[test_mask][target],test_pred_lr))
print("\nGINI Train:",round(2*roc_auc_score(data[train_mask][target],train_pred_lr)-1,4)*100)
print("GINI Test:",round(2*roc_auc_score(data[test_mask][target],test_pred_lr)-1,4)*100)

print("\nSelected Features:")
print(selected_features)
```
