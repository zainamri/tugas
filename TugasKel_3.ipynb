```python
#1.	Rumah makan padang memiliki menu yang enak. Mereka ingin memotong beberapa item dari menu. Untuk membuat sebagian besar pelanggan mereka senang, mereka ingin menghapus hanya 3 menu makanan yang paling tidak populer dari menu. Kita memiliki menu makanan yang disajikan dalam rumah makan tersebut, serta hitungan berapa kali makanan tersebut terjual dalam seminggu terakhir. buat visualisasi yang menunjukan popularitas relatif dari menu makanan tersebut. Sebutkan menu makanan yang harus disingkirkan. Data dapat dilihat dibawah ini!

masakan_padang = {
  'Rendang':95,
  'Dendeng':93,
  'Ayam Bakar':90,
  'Ayam Gulai':92,
  'Gulai Tunjang':90,
  'Gulai Pakis':83,
  'Kikil':85,
  'Ikan Tongkol Balado':89,
  'Telor Balado':87,
  'Perkedel':84
}

```

```python
import matplotlib.pyplot as plt

masakan_padang = {
  'Rendang':95,
  'Dendeng':93,
  'Ayam Bakar':90,
  'Ayam Gulai':92,
  'Gulai Tunjang':90,
  'Gulai Pakis':83,
  'Kikil':85,
  'Ikan Tongkol Balado':89,
  'Telor Balado':87,
  'Perkedel':84
}


# membuat dan menampilkan chart pie
values = masakan_padang.values()
labels = masakan_padang.keys()

plt.pie(values, labels=labels, autopct='%1.1f%%')
plt.title('Popularitas Menu Masakan Padang')
plt.show()
```

1. Chart apa yang anda pilih untuk problem diatas dan mengapa anda memilih chart tersebut?
~ Chart yang dipilih ialah Pie Chart
~ alasan menggunakan pie chart karena:
  1. Data hanya berisi nama masakan dan populasi, cocok untuk menggambarkan proporsi
  2. Soal meminta masakan paling tidak populernya, pie chart mudah menunjukkan bagian terkecil
  3. Tampilan pie chart rapi
  4. Mudah mengidentifikasi masakan terkecil sebagai jawaban soal

```python
# menentukan nilai minimum populasi

min_pop_3 = sorted(masakan_padang.values())[2]

menu_hapus = [k for k,v in masakan_padang.items() if v <= min_pop_3]
print('2. Menu Makanan apa saja yang perlu anda sarankan untuk dikeluarkan dari menu?')
print("Menu yang disarankan untuk dihapus/ dikeluarkan:",menu_hapus)
```

```python
#2.	Kedai Salgits memiliki berbagai pilihan menu camilan. Untuk menyesuaikan selera konsumen, manajemen ingin mengurangi 3 menu yang kurang laris. Berikut data penjualan menu dalam seminggu terakhir:
Menu_salgits = {
    'Salad Buah':65,
    'Salad Sayur':60,
    'Cireng Ayam':70,
    'Cireng Kornet':67,
    'Cireng Keju':65,
    'Cireng Sosis':58,
    'Cireng Bakso':56,
    'JASUKE':55,
    'Corndog Manis':61,
    'Corndog Asin':59
}

# Buatlah visualisasi untuk mengetahui popularitas relatif setiap menu. Sebutkan 3 menu yang perlu dihapus berdasarkan data tersebut.
```

```python
import matplotlib.pyplot as plt

Menu_salgits = {
    'Salad Buah':65,
    'Salad Sayur':60,
    'Cireng Ayam':70,
    'Cireng Kornet':67,
    'Cireng Keju':65,
    'Cireng Sosis':58,
    'Cireng Bakso':56,
    'JASUKE':55,
    'Corndog Manis':61,
    'Corndog Asin':59
}

x = list(Menu_salgits.values())
labels = list(Menu_salgits.keys())

plt.barh(labels, x)
plt.title('Popularitas Menu di Kedai Salgits')
plt.ylabel('Jenis Menu')
plt.xlabel('Jumlah Penjualan')
plt.show()
```

1. .   Chart apa yang anda pilih untuk problem diatas dan mengapa anda memilih chart tersebut?
~ Chart yang dipilih ialah Horizontal Bar Chart
~ alasan menggunakan Horizontal bar chart karena:
   1. Mudah membandingkan panjang bar setiap menu secara visual
   2. Dengan melihat 3 bar terpendek, mudah mengidentifikasi 3 menu terendah
   3. Tampilannya lebih kompak daripada bar chart vertikal
   4. Tetap mudah dibaca meski jumlah data menu banyak

```python
# mengambil 3 penjualan terendah
min_sales_3 = sorted(Menu_salgits.values())[2]

menus_to_remove = [menu for menu,sales in Menu_salgits.items() if sales <= min_sales_3]
print('2. Menu Makanan apa saja yang perlu anda sarankan untuk dikeluarkan dari menu?')
print("Menu yang disarankan untuk dihapus/ dikeluarkan:")
print(menus_to_remove)
```
