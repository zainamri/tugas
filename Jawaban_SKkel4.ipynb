```python
#Kasus 1:
#Sebuah restoran ingin melakukan studi tentang preferensi makanan dari pelanggan mereka.
#Mereka ingin memahami jenis makanan apa yang paling diminati oleh pelanggan mereka.
#Dalam studi ini, kita akan menggunakan dua atribut label: jenis makanan (misalnya, pizza, burger, pasta, salad, dll.) dan jumlah pelanggan yang memilih setiap jenis makanan tersebut.

Data: {
      "Pizza:", 30,
      "Burger:", 25,
       "Pasta:", 20,
       "Salad:", 15,
       "Steak:", 12,
       "Sushi:", 10,
       "Chicken Wings:", 8,
       "Tacos:", 7,
       "Nasi Goreng:", 6,
       "Kebab:", 5,
       "Fish and Chips:", 4,
       "Ramen:", 3,
       "Dim Sum:", 3,
       "Satay:", 2,
       "Sate Padang:", 1
}
```

```python
#Jawab
import matplotlib.pyplot as plt

data = {"Pizza": 30,
        "Burger": 25,
        "Pasta": 20,
        "Salad": 15,
        "Steak": 12,
        "Sushi": 10,
        "Chicken Wings": 8,
        "Tacos": 7,
        "Nasi Goreng": 6,
        "Kebab": 5,
        "Fish and Chips": 4,
        "Ramen": 3,
        "Dim Sum": 3,
        "Satay": 2,
        "Sate Padang": 1}

labels = list(data.keys())
sizes = list(data.values())

fig, ax = plt.subplots()

ax.barh(labels, sizes, color='#FF9933')
ax.invert_yaxis()

plt.title('Makanan Terfavorit')
plt.show()
```

1. .   Chart apa yang anda pilih untuk problem diatas dan mengapa anda memilih chart tersebut?
*    Chart yang dipilih ialah Horizontal Bar Chart
*    Alasan menggunakan bar chart horizontal dengan warna jingga:
    1. Bar chart baik untuk melihat perbandingan jumlah data
    2. Bentuknya berbeda dari pie chart
    3. Warna jingga menarik perhatian dan memberi highlight
    4. Warna jingga membuat visualisas terlihat lebih ceria dan segar
    5. Cocok mewakili data makanan gaya hidup santai

```python
#soal 2, mengambil 3 penjualan terlaris
item_3 = sorted(data.values())[-3]

menu_to_recom = [menu for menu,sales in data.items() if sales >= item_3]
print('2. menu makan apa saja yang direkomendasikan ?')
print("Menu yang disarankan untuk direkomendasikan:")
print(menu_to_recom)
```

```python
#Jawaban Kasus 2
# code untuk chart
import matplotlib.pyplot as plt

jenis = {"Elektronik": 120,
         "Fashion": 90,
         "Perabotan": 80,
         "Mainan": 60,
         "Makanan": 110,
         "Kosmetik": 70,
         "Buku": 45,
         "Alat Olahraga": 55,
         "Perhiasan": 65,
         "Elektronik": 130,
         "Fashion": 95,
         "Perabotan": 85,
         "Mainan": 70,
         "Makanan": 120,
         "Kosmetik": 75}

sizes = jenis.values()
labels = jenis.keys()

fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, radius=1.0, wedgeprops={'edgecolor':'w'}, startangle=90,
       autopct='%1.0f%%')

centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.axes[0].add_artist(centre_circle)

plt.title("Grafik Donut Penjualan Berbagai Jenis Barang")
plt.show()

```
